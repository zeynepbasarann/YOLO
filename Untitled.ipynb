```python
import cv2
from ultralytics import YOLO

model = YOLO('yolo11n-pose.pt')  


kamera = cv2.VideoCapture(0)  

if not kamera.isOpened():
    print("Kamera balatlamad!")
    exit()

while True:
    ret, frame = kamera.read()
    if not ret:
        print("Kamera grnts alnamad!")
        break


    results = model(frame)


    annotated_frame = results[0].plot()


    cv2.imshow("YOLOv11 Pose Tespiti", annotated_frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


kamera.release()
cv2.destroyAllWindows()
kamerada yoloposealglyor
```
